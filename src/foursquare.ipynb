```python
import requests
import json
import pandas as pd
from getpass import getpass
import os
from dotenv import load_dotenv
```

```python
def foursquare_data(data):
    rows = []
    for result in data['results']:
        row = {
            'name': result['name'],
            'distance': result['distance'],
            'latitude': result['geocodes']['main']['latitude'],
            'longitude': result['geocodes']['main']['longitude'],
            'category_id': result['categories'][0]['id'],
            'category_name': result['categories'][0]['name']
        }
        rows.append(row)
    df = pd.DataFrame(rows).sort_values(by='distance', ascending=True)
    return df
```

```python
token = os.getenv("token")
load_dotenv()
```

```python
url_airport =  'https://api.foursquare.com/v3/places/search?ll=37.7749%2C-122.4194&radius=25000&categories=19040' \

headers = {
    "accept": "application/json",
    "Authorization": token }
response_airports = requests.get(url_airport, headers=headers)
```

```python
response_airports
```

```python
airports_raw = response_airports.json()
airports = foursquare_data(airports_raw)
airports
```
